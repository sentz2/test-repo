# Random Numbers Example
This notebook generates random numbers, computes the average, and plots them.

```python
import random
import matplotlib.pyplot as plt
```

```python
# Generate 20 random numbers
numbers = [random.randint(1, 100) for _ in range(20)]
numbers
```

```python
# Compute average
average = sum(numbers) / len(numbers)
average
```

```python
# Plot numbers and average
plt.plot(numbers, marker="o")
plt.axhline(average, color="red", linestyle="--", label=f"Average = {average:.2f}")
plt.legend()
plt.title("Random Numbers and Their Average")
plt.show()
```
